### Exploration


```python
base_color=sns.color_palette()[0]
plt.figure(figsize=[20,10])
sns.countplot(data=df,x='Education_LVL',color=base_color)
plt.title('The Number of persons in each Educational_LVL',
          fontweight='bold', fontsize=16)
plt.xlabel('Education_lvl', fontweight='bold', fontsize=12)
plt.ylabel('Counts',fontweight='bold', fontsize=12);
```

```python
base_color=sns.color_palette()[0]
plt.figure(figsize=[20,10])
sns.countplot(data=df,x='Learning_Practice',color=base_color)
plt.title('A Count on the number of learning practices',
          fontweight='bold', fontsize=16)
plt.xlabel('Learning_Practice', fontweight='bold', fontsize=12)
plt.ylabel('Counts',fontweight='bold', fontsize=12);
```

```python
base_color=sns.color_palette()[0]
plt.figure(figsize=[20,10])
sns.countplot(data=df,x='Mentorship_Guide',color=base_color)
plt.title('A count on the ratings on the availability of Mentor/Guide present 24/7',
          fontweight='bold', fontsize=16)
plt.xlabel('Mentor_Rating', fontweight='bold', fontsize=12)
plt.ylabel('Counts',fontweight='bold', fontsize=12);
```

```python
plt.figure(figsize=[12,10])
df['Pace_Learning'].value_counts().plot(kind='pie',autopct='%.2f')
plt.title('The Number of persons who enjoy learning at their own pace',
          fontweight='bold', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=[12,10])
df['Mentorship_Importance'].value_counts().plot(kind='pie',autopct='%.2f')
plt.title('The Number of persons who think mentorship is important ',
          fontweight='bold', fontsize=16)
plt.show()
```
